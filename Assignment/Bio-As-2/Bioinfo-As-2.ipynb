```python
cluster_1_data = filtered_data.loc[filtered_data.index.isin(cell_clusters[cell_clusters['Cluster'] == top_clusters[0]]["Cell"])]
cluster_2_data = filtered_data.loc[filtered_data.index.isin(cell_clusters[cell_clusters['Cluster'] == top_clusters[1]]["Cell"])]
```

```python
from scipy.stats import ttest_ind
import numpy as np
import pandas as pd

# åˆå§‹åŒ–å­˜å‚¨ç»“æžœçš„åˆ—è¡¨
genes = filtered_data.columns
pvalues = []
fold_changes = []

# å¯¹æ¯ä¸ªåŸºå› è®¡ç®— P å€¼å’Œ Fold Change
for gene in genes:
    group1 = cluster_1_data[gene]
    group2 = cluster_2_data[gene]
    
    # T æ£€éªŒ
    t_stat, p_value = ttest_ind(group1, group2, equal_var=False)
    pvalues.append(p_value)
    
    # Fold Change
    mean_group1 = np.mean(group1)
    mean_group2 = np.mean(group2)
    fold_change = mean_group1 / mean_group2 if mean_group2 != 0 else np.inf
    fold_changes.append(fold_change)

# åˆ›å»ºç»“æžœ DataFrame
diff_genes = pd.DataFrame({
    'Gene': genes,
    'PValue': pvalues,
    'FoldChange': fold_changes
})
```

```python
from statsmodels.stats.multitest import multipletests

# è°ƒæ•´ P å€¼
adjusted_pvalues = multipletests(diff_genes['PValue'], alpha=0.05, method='fdr_bh')[1]
diff_genes['AdjustedPValue'] = adjusted_pvalues

# ç­›é€‰æ˜¾è‘—å·®å¼‚è¡¨è¾¾åŸºå› 
significant_genes = diff_genes[
    (diff_genes['AdjustedPValue'] < 0.05) & (diff_genes['FoldChange'] > 2)
]

# æ‰“å°ç»“æžœ
print(f"æ£€æµ‹åˆ° {len(significant_genes)} ä¸ªæ˜¾è‘—å·®å¼‚è¡¨è¾¾åŸºå› ")

# ä¿å­˜ç»“æžœ
diff_genes.to_csv('differential_genes_full.csv', index=False)
significant_genes.to_csv('significant_differential_genes.csv', index=False)
```

```python
import pandas as pd
import statsmodels.api as sm
from scipy.stats import norm

# è¾“å…¥æ•°æ®
data = {
    "Gene 1": [4.3, 5.6, 4.4, 2.6, 3.5, 2.6],
    "Gene 2": [3.3, 3.5, 1.7, 3.5, 4.1, 3.8],
    "Gene 3": [2.4, 3.7, 4.7, 5.6, 5.8, 4.8],
    "Gene 4": [5.3, 5.6, 4.4, 0.4, 1.1, 2.6],
    "Gene 5": [3.3, 2.6, 2.5, 4.6, 3.9, 4.6],
    "Gene 6": [2.3, 3.3, 2.4, 5.2, 4.5, 5.6],
}
group = [1, 1, 1, 0, 0, 0]  # T=1, C=0

df = pd.DataFrame(data)
results = []

# å‡è®¾ b1 ~ N(0, 1)
alpha = 0.1
critical_value = norm.ppf(1 - alpha / 2)

for gene in df.columns:
    y = df[gene]
    x = sm.add_constant(group)
    
    # æ‹Ÿåˆçº¿æ€§æ¨¡åž‹
    model = sm.OLS(y, x).fit()
    b0, b1 = model.params
    p_value = model.pvalues[1]
    
    # åˆ¤æ–­æ˜¾è‘—æ€§
    significant = abs(b1) > critical_value
    results.append({"Gene": gene, "b0": b0, "b1": b1, "Significant": significant})

# è¾“å‡ºç»“æžœ
df_results = pd.DataFrame(results)
print(df_results)
```

```python
import pandas as pd
from sklearn.metrics import mutual_info_score
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt


data = {
    "G1": [5, 4, 4, 5, 3, 6],
    "G2": [2, 3, 3, 7, 2, 5],
    "G3": [3, 2, 3, 4, 3, 6],
    "G4": [3, 3, 3, 6, 1, 3]
}
df = pd.DataFrame(data)


genes = df.columns
n_genes = len(genes)
mi_matrix = np.zeros((n_genes, n_genes))

for i in range(n_genes):
    for j in range(n_genes):
        if i != j:
            mi_matrix[i, j] = mutual_info_score(df[genes[i]], df[genes[j]])


clr_matrix = np.zeros_like(mi_matrix)
for i in range(n_genes):
    for j in range(n_genes):
        if i != j:
            z_i = (mi_matrix[i, j] - np.mean(mi_matrix[i, :])) / np.std(mi_matrix[i, :])
            z_j = (mi_matrix[i, j] - np.mean(mi_matrix[:, j])) / np.std(mi_matrix[:, j])
            clr_matrix[i, j] = np.sqrt(z_i**2 + z_j**2)


threshold = 0.1
edges = []
for i in range(n_genes):
    for j in range(i + 1, n_genes):
        if clr_matrix[i, j] > threshold:
            edges.append((genes[i], genes[j], clr_matrix[i, j]))


G = nx.Graph()
for edge in edges:
    G.add_edge(edge[0], edge[1], weight=edge[2])

pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=2000, font_size=12)
nx.draw_networkx_edge_labels(G, pos, edge_labels={(u, v): f"{d['weight']:.2f}" for u, v, d in G.edges(data=True)})
plt.title('Inferred Gene Network (CLR > 0.1)')
plt.show()
```
